```python
# NOME: DeividSantos
# As repostas das perguntas estão no arquivo respostas

#https://www.kaggle.com/code/italomarcelo/nlp-spacy-portuguese

import spacy.cli
import os
import nltk

from polyglot.text import Text
from nltk.stem.porter import *

spacy.cli.download("pt_core_news_sm")
stemmer = PorterStemmer()

def ler_texto_arquivo(caminho_arquivo):
    with open(caminho_arquivo, 'r', encoding='UTF-16 LE') as arquivo:
        return arquivo.read()

def escrever_texto_arquivo(caminho_arquivo, texto):
    with open(caminho_arquivo, 'w') as arquivo:
        arquivo.write(texto)

def criar_arquivo_tokenizado(doc):
    novo_doc = ""
    for token in doc:
        novo_doc = novo_doc + " " + token.text
    caminho_novo_arquivo = os.path.join("TOKENIZADO", nome_arquivo)
    escrever_texto_arquivo(caminho_novo_arquivo, novo_doc)
    
def criar_arquivo_lemma(doc):
    novo_doc = ""
    for token in doc:
        novo_doc = novo_doc + token.text + ", " + token.lemma_ + "\n"
    caminho_novo_arquivo = os.path.join("LEMMA", nome_arquivo)
    escrever_texto_arquivo(caminho_novo_arquivo, novo_doc)
    
def criar_arquivo_pos(doc):
    novo_doc = ""
    for token in doc:
        novo_doc = novo_doc + token.text + ", " + token.pos_ + "\n"
    caminho_novo_arquivo = os.path.join("POS", nome_arquivo)
    escrever_texto_arquivo(caminho_novo_arquivo, novo_doc)

def criar_arquivo_stemming(doc):
    novo_doc = ""
    for token in doc:
        novo_doc = novo_doc + token.text + ", " + stemmer.stem(token.text) + "\n"
    caminho_novo_arquivo = os.path.join("STEMMING", nome_arquivo)
    escrever_texto_arquivo(caminho_novo_arquivo, novo_doc)

def criar_arquivo_sem_stop_words(doc):
    novo_doc = ""
    for token in doc:
        if not(token.is_stop):
            novo_doc = novo_doc + " " + token.text
    caminho_novo_arquivo = os.path.join("SEM_STOP_WORDS", nome_arquivo)
    escrever_texto_arquivo(caminho_novo_arquivo, novo_doc)

def criar_arquivo_completo(doc):
    novo_doc = "TOKEN, STEMMING, LEMMA, POS, IS_STOP_WORD \n"
    for token in doc:
        novo_doc = novo_doc + token.text + ", " + stemmer.stem(token.text)+ ", " +  token.lemma_  + ", " + token.pos_  + ", " + str(token.is_stop) +"\n"
    caminho_novo_arquivo = os.path.join("TODOS", nome_arquivo)
    escrever_texto_arquivo(caminho_novo_arquivo, novo_doc)

    
#Carregando o pacote e o texto
nlp = spacy.load("pt_core_news_sm")
pasta = "DADOS_FOLHADESAOPAULO"
arquivos = os.listdir(pasta)
stopwords = nlp.Defaults.stop_words

for nome_arquivo in arquivos:
    caminho_arquivo = os.path.join(pasta, nome_arquivo)
    arquivo = ler_texto_arquivo(caminho_arquivo)
    doc = nlp(arquivo)
    criar_arquivo_tokenizado(doc)
    criar_arquivo_lemma(doc)
    criar_arquivo_pos(doc)
    criar_arquivo_stemming(doc)
    criar_arquivo_sem_stop_words(doc)
    criar_arquivo_completo(doc)



frase = "O ponta Rivaldo não interessa mais ao Corinthians. O diretor de futebol Carlos Nujud teve uma reunião com o presidente do Mogi Mirim, Wilson de Barros, e deu como finalizada a negociação. \"Eles não abaixaram a pedida: US$ 4,5 milhões\", explicou o dirigente corintiano. \nQuanto ao meia Dener, a diretoria deve apresentar a sua contratação hoje. O Conselho de Orientação e Fiscalização da Portuguesa deveria fixar o preço do passe do jogador até ontem à noite. O clube ofereceu US$ 2 milhões, mais os passes de dois jogadores –entre uma lista apresentada. O Corinthians iniciou ontem a pré-temporada em Itamonte. O time fica concentrado até terça."

texto = Text(frase)
marcacoes_pos = texto.pos_tags

for palavra, tag_pos in marcacoes_pos:
    print(palavra + ", " + tag_pos)

print("Finalizado") 
```
