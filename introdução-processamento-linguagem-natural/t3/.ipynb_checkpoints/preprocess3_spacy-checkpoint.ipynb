```python
def ler_texto_arquivo(caminho_arquivo):
    with open(caminho_arquivo, 'r', encoding='UTF-16 LE') as arquivo:
        return arquivo.read()

#Carregando o pacote e o texto
nlp = spacy.load("pt_core_news_sm")
pasta = "DADOS_FOLHADESAOPAULO"
arquivos = os.listdir(pasta)
stopwords = nlp.Defaults.stop_words

i = 0
for nome_arquivo in arquivos[:5]:
    caminho_arquivo = os.path.join(pasta, nome_arquivo)
    arquivo = ler_texto_arquivo(caminho_arquivo)
    doc = nlp(arquivo)
    for token in doc:
        if token.pos_ == "NOUN" or token.pos_ == "PROPN":
            print(token.text + ", " + token.pos_ + ", " + nome_arquivo)
```
